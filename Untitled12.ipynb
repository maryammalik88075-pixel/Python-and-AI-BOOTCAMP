```python
# STEP 1: Install required packages
!pip install -q scikit-learn matplotlib seaborn plotly

# STEP 2: Import libraries
import os
import zipfile
import warnings
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    confusion_matrix, classification_report, roc_curve, auc
)
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.impute import SimpleImputer
from google.colab import files

# STEP 3: Suppress warnings
warnings.filterwarnings("ignore")

# STEP 4: Upload and unzip dataset
uploaded = files.upload()
zip_path = next(iter(uploaded))
extract_dir = "/content/data"
os.makedirs(extract_dir, exist_ok=True)
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_dir)

# STEP 5: Load and aggregate data by file
csv_files = []
for root, _, files_in_dir in os.walk(extract_dir):
    for file in files_in_dir:
        if file.endswith(".csv"):
            csv_files.append(os.path.join(root, file))

print(f"\nüîç Found {len(csv_files)} CSV files.")

aggregated_dfs = []
for file in csv_files:
    try:
        df_temp = pd.read_csv(file)
        if "nonsepsis" in file.lower():
            df_temp["label"] = 0
        elif "sepsis" in file.lower():
            df_temp["label"] = 1
        # Aggregate by taking the mean of numeric columns for the entire file
        aggregated_df = pd.DataFrame(df_temp.mean(numeric_only=True)).T  # Transpose to row
        aggregated_df["label"] = df_temp["label"].iloc[0]  # Keep the first label
        aggregated_dfs.append(aggregated_df)
    except Exception as e:
        print(f"Error loading {file}: {e}, skipping...")
        continue

if not aggregated_dfs:
    raise ValueError("‚ùå No data loaded after aggregation!")

df = pd.concat(aggregated_dfs, ignore_index=True)
print(f"\n‚úÖ Loaded data (aggregated): {df.shape}")
print("Label column exists:", "label" in df.columns)

# STEP 6: Data Handling - Initial checks and cleaning
print("\n=== DATA HANDLING ===")
print("Initial dataset info:")
print(df.info())
print("\nMissing values per column:")
print(df.isnull().sum())
print("\nDuplicate rows:", df.duplicated().sum())

# Handle missing values
df.dropna(axis=1, how='all', inplace=True)  # Drop columns with all missing values
# Fill remaining missing values with column means
imputer = SimpleImputer(strategy='mean')
df_imputed = pd.DataFrame(imputer.fit_transform(df.select_dtypes(include=[np.number])),
                         columns=df.select_dtypes(include=[np.number]).columns)
df[df_imputed.columns] = df_imputed

# Handle duplicates
df.drop_duplicates(inplace=True)
print("Duplicate rows after cleaning:", df.duplicated().sum())

# Convert categorical features if any
categorical_cols = df.select_dtypes(include=['object']).columns
if len(categorical_cols) > 0:
    print(f"\nConverting {len(categorical_cols)} categorical features to numerical")
    for col in categorical_cols:
        df[col] = LabelEncoder().fit_transform(df[col].astype(str))

# STEP 7: Exploratory Data Analysis (EDA)
print("\n=== EXPLORATORY DATA ANALYSIS ===")
print("Basic statistics:")
print(df.describe())

# Check class distribution
class_dist = df['label'].value_counts()
print(f"\nClass distribution:\n{class_dist}")

# Visualizations with Matplotlib/Seaborn
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.countplot(x='label', data=df)
plt.title('Class Distribution')

plt.subplot(1, 2, 2)
# Select a few important features for correlation visualization
if len(df.columns) > 5:
    numeric_cols = df.select_dtypes(include=[np.number]).columns
    # Select up to 10 features with highest correlation to label
    correlations = df[numeric_cols].corrwith(df['label']).abs().sort_values(ascending=False)
    top_features = correlations.index[1:6]  # Skip label itself
    sns.heatmap(df[top_features].corr(), annot=True, cmap='coolwarm')
    plt.title('Correlation Matrix of Top Features')
else:
    sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
    plt.title('Correlation Matrix')

plt.tight_layout()
plt.show()

# Interactive plot with Plotly
if len(df.columns) > 2:
    # Select two features with highest correlation to label for scatter plot
    numeric_cols = df.select_dtypes(include=[np.number]).columns
    correlations = df[numeric_cols].corrwith(df['label']).abs().sort_values(ascending=False)
    top_two = correlations.index[1:3]  # Skip label itself

    fig = px.scatter(df, x=top_two[0], y=top_two[1], color='label',
                     title=f'Interactive Scatter Plot: {top_two[0]} vs {top_two[1]}',
                     hover_data=df.columns)
    fig.show()

# STEP 8: Feature Engineering
print("\n=== FEATURE ENGINEERING ===")
X = df.drop("label", axis=1)
y = df["label"]

# Normalize/scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)

# Perform train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)

print(f"Train set size: {X_train.shape[0]}, Test set size: {X_test.shape[0]}")
print("Test set class distribution:")
print(y_test.value_counts())

# STEP 9: Model Training (Baseline)
print("\n=== MODEL TRAINING (BASELINE) ===")
models = {
    "K-Nearest Neighbors": KNeighborsClassifier(),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42)
}

baseline_results = []

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    baseline_results.append({
        "Model": name,
        "Accuracy": accuracy_score(y_test, y_pred),
        "Precision": precision_score(y_test, y_pred),
        "Recall": recall_score(y_test, y_pred),
        "F1 Score": f1_score(y_test, y_pred)
    })

baseline_df = pd.DataFrame(baseline_results)
print("\nBaseline Model Performance:")
display(baseline_df)

# STEP 10: Feature Importance (Random Forest)
print("\n=== FEATURE IMPORTANCE ===")
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)

# Get feature importances
importances = rf_model.feature_importances_
feature_importance_df = pd.DataFrame({
    'Feature': X.columns,
    'Importance': importances
}).sort_values('Importance', ascending=False)

# Plot feature importances
plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=feature_importance_df.head(10))
plt.title('Top 10 Feature Importances (Random Forest)')
plt.tight_layout()
plt.show()

print("Top 10 most important features:")
print(feature_importance_df.head(10))

# STEP 11: Hyperparameter Tuning
print("\n=== HYPERPARAMETER TUNING ===")
# Define parameter distributions for each model
param_distributions = {
    "K-Nearest Neighbors": {
        'n_neighbors': range(1, 31),
        'weights': ['uniform', 'distance'],
        'metric': ['euclidean', 'manhattan', 'minkowski']
    },
    "Decision Tree": {
        'max_depth': [None, 10, 20, 30, 50, 100],
        'min_samples_split': range(2, 11),
        'min_samples_leaf': range(1, 11)
    },
    "Random Forest": {
        'n_estimators': [50, 100, 200, 300],
        'max_depth': [None, 10, 20, 30],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4]
    }
}

tuned_models = {}
tuned_results = []

for name in models.keys():
    print(f"\nTuning {name}...")
    model = models[name]

    # Perform randomized search
    random_search = RandomizedSearchCV(
        estimator=model,
        param_distributions=param_distributions[name],
        n_iter=20,
        cv=5,
        scoring='accuracy',
        random_state=42,
        n_jobs=-1
    )

    random_search.fit(X_train, y_train)
    tuned_models[name] = random_search.best_estimator_

    # Evaluate tuned model
    y_pred = random_search.predict(X_test)

    tuned_results.append({
        "Model": f"{name} (Tuned)",
        "Accuracy": accuracy_score(y_test, y_pred),
        "Precision": precision_score(y_test, y_pred),
        "Recall": recall_score(y_test, y_pred),
        "F1 Score": f1_score(y_test, y_pred),
        "Best Params": str(random_search.best_params_)
    })

    print(f"Best parameters for {name}: {random_search.best_params_}")

# Combine baseline and tuned results
all_results = pd.concat([baseline_df, pd.DataFrame(tuned_results).drop('Best Params', axis=1)])
print("\nComparison of Baseline vs Tuned Models:")
display(all_results)

# STEP 12: Model Evaluation
print("\n=== MODEL EVALUATION ===")
# Find the best model based on accuracy
best_model_name = all_results.iloc[all_results['Accuracy'].idxmax()]['Model']
print(f"Best performing model: {best_model_name}")

# Get the best model
if "(Tuned)" in best_model_name:
    base_name = best_model_name.replace(" (Tuned)", "")
    best_model = tuned_models[base_name]
else:
    best_model = models[best_model_name]

# Generate predictions with the best model
y_pred = best_model.predict(X_test)
y_prob = best_model.predict_proba(X_test)[:, 1] if hasattr(best_model, "predict_proba") else None

# Print classification report
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=['Non-Sepsis', 'Sepsis']))

# Confusion Matrix
plt.figure(figsize=(6, 5))
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Non-Sepsis', 'Sepsis'],
            yticklabels=['Non-Sepsis', 'Sepsis'])
plt.title(f'Confusion Matrix - {best_model_name}')
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()

# ROC Curve (if the model supports probability predictions)
if y_prob is not None:
    fpr, tpr, _ = roc_curve(y_test, y_prob)
    roc_auc = auc(fpr, tpr)

    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'ROC Curve - {best_model_name}')
    plt.legend(loc="lower right")
    plt.show()
else:
    print("ROC curve not available for this model type")

# STEP 13: Conclusion
print("\n=== CONCLUSION ===")
print("Based on the analysis:")
print(f"1. The best performing model is {best_model_name} with an accuracy of {all_results.loc[all_results['Model'] == best_model_name, 'Accuracy'].values[0]:.3f}")
print("2. The most important features are:")
for i, row in feature_importance_df.head(5).iterrows():
    print(f"   - {row['Feature']} (Importance: {row['Importance']:.4f})")

# Check if tuning improved results
improvements = []
for model_name in models.keys():
    baseline_acc = baseline_df.loc[baseline_df['Model'] == model_name, 'Accuracy'].values[0]
    tuned_acc = all_results.loc[all_results['Model'] == f"{model_name} (Tuned)", 'Accuracy'].values[0]
    improvement = tuned_acc - baseline_acc
    improvements.append((model_name, improvement))

print("3. Hyperparameter tuning improvements:")
for model_name, improvement in improvements:
    print(f"   - {model_name}: {improvement:.4f} accuracy improvement")

print("\nPipeline execution completed! ‚úÖ")
```
